```python
# Import libraries

import pandas as pd
import os
from darwinex_ftp import DarwinexFtp
```

```python
# Indicate user parameters for the FTP connection

ftp_server = 'darwindata.darwinex.com'
port = 21
username = 'your_Darwinex_username'
password = 'your_password'
```

```python
# Initialize DarwinexFtp object and connect to the FTP server

dwf = DarwinexFtp(ftp_server=ftp_server, port=port, username=username, password=password)
```

```python
# Get a list of available darwins

ls_darwin = dwf.get_ftp_files()
print(ls_darwin)
```

```python
# Get the available dates for a given list of darwins

ls_assets=['THA', 'JTL']
df_dates = dwf.get_darwins_dates(ls_assets=ls_assets)
df_dates
```

```python
# Download quotes data of a given list of darwins

# Indicate the list of darwins to download
ls_assets = ['THA', 'JTL']  
# Indicate if you want to download the former var10 data (added as additional columns) 
include_var10 = True        
# Indicate the start period for the data to download
start_period='2023-01'      
# Indicate the end period for the data to download
end_period='2023-03'        
# If you want the data to be resampled, indicate resample frequency using Pandas rule. E.g. resample = 'D'. If not, indicate resample = None
resample = 'B'              
# Path to store a pickle containing the df_quotes data
    # If a df_quotes.pkl file already exists in the indicated path, it will be loaded at the beginning, and the new downloaded data will be merged with the existing one
    # If you do not want to save the data, indicate path = None
path = 'C:\\Users\\username\\Documents'
path = 'E:\\github\\darwinex-data'
            
# Display information about the download process. 0: no information, 1-3: information
verbose = 1                 

# Download the data
df_quotes = dwf.download_ftp_quotes_data(ls_assets=ls_assets, include_var10=include_var10, start_period=start_period, end_period=end_period, resample=resample, path=path, verbose=verbose)
```

```python
# Close FTP connection. It will be closed automatically after 30 minutes of inactivity

dwf.close_ftp_connection()
```

```python
# Load df_quotes from pickle file

path = 'C:\\Users\\your_user_name\\Desktop'
file_name = 'quotes_data.pkl'
file_path = path + os.sep + file_name
df_quotes = pd.read_pickle(file_path)
```
